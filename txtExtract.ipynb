```python
import pyautogui as pg, pyperclip as pc, time, re, pandas as pd
y = 250
```

```python
data = []
codigo = []
historico = []
valor = []
for m in range(12):
    if m == 0: 
        y = 230
        ip = 9
    else: 
        y = 250
        ip = 8
    if m+1 >= 10: mes = str(m+1) 
    else: mes = '0' + str(m+1)
    pg.click(2200+m*36, 150)
    time.sleep(0.2)
    for l in range(10000000):
        if 30*l > 30*ip:
            for scroll in range(l-ip):
                    pg.click(3460, 500)
            pg.doubleClick(2300, y+30*ip)
        else:
            pg.doubleClick(2300, y+30*l)
            
        #dentro
        pg.click(2220, 180)
        pg.hotkey('ctrl', 'a')
        pg.hotkey('ctrl', 'c')
        time.sleep(0.3)
        data.append(pc.paste())
        if len(valor)>1:
            if data[-1] == valor[-1]:
                data.pop()
                codigo.pop()
                historico.pop()
                valor.pop()
                data.pop()
                print('quebrou aqui')
                break
        pg.click(2220, 230)
        time.sleep(0.3)
        pg.hotkey('ctrl', 'c')
        codigo.append(pc.paste())
        pg.click(2220, 270)
        pg.hotkey('ctrl', 'a')
        pg.hotkey('ctrl', 'c')
        historico.append(pc.paste())
        pg.click(2220, 390)
        time.sleep(0.5)
        pg.hotkey('ctrl', 'c')
        valor.append(pc.paste())
        pg.click(3350, 830)
        if len(valor)>1:
            if 'venda' in historico[-1]:
                valor[-1] = valor[-1]*-1
            if valor[-1] == valor[-2] and data[-1] == data[-2] and historico[-1] == historico[-2]:
                data.pop()
                codigo.pop()
                historico.pop()
                valor.pop()
                print('quebrou aqui2')
                break
print(data)
print(codigo)
print(historico)
print(valor)
```
